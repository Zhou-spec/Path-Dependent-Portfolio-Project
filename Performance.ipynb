```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
%matplotlib ipympl

year_2020 = yf.download('SPY', start='2020-01-01', end='2022-12-31')
year_2020['Adj Close']
numpy_year_2020 = year_2020['Adj Close'].to_numpy()
numpy_61_91_2020 = numpy_year_2020[61: 91]
numpy_361_391_2020 = numpy_year_2020[361: 391]
numpy_521_551_2020 = numpy_year_2020[521: 551]
y_61_91 = (numpy_61_91_2020[1:]-numpy_61_91_2020[:-1]) / numpy_61_91_2020[:-1]
y_361_391 = (numpy_361_391_2020[1:]-numpy_361_391_2020[:-1]) / numpy_361_391_2020[:-1]
y_521_551 = (numpy_521_551_2020[1:]-numpy_521_551_2020[:-1]) / numpy_521_551_2020[:-1]

wealth = 1.0
return_61_91 = [1]
return_361_391 = [1]
return_521_551 = [1]

wealth = 1
for i in range(len(y_61_91)):
    wealth = wealth * (1+y_61_91[i])
    return_61_91.append(wealth)

wealth = 1
for i in range(len(y_361_391)):
    wealth = wealth * (1 + y_361_391[i])
    return_361_391.append(wealth)

wealth = 1
for i in range(len(y_521_551)):
    wealth = wealth * (1 + y_521_551[i])
    return_521_551.append(wealth)


return_61_91 = np.array(return_61_91)
return_361_391 = np.array(return_361_391)
return_521_551 = np.array(return_521_551)


x_61_91 = np.load('61-91-trading-days.npy')
x_361_391 = np.load('361-391-trading-days.npy')
x_521_551 = np.load('521-551-trading-days.npy')
```

```python
plt.figure()
plt.plot(x_61_91, label='Portfolio', color = 'blue')
plt.plot(return_61_91 / return_61_91[0], label='S&P 500', color = 'red')
plt.show()
plt.savefig('61-91-trading-days.png')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(return_361_391 / return_361_391[0], label='S&P 500', color = 'red')
plt.plot(x_361_391, label='Portfolio', color = 'blue')
plt.show()
plt.savefig('361-391-trading-days.png')
```

```python
plt.figure(figsize = (10, 6))
plt.xlabel('528-to-551 trading days from 2020-01-01')
plt.ylabel('Cumulative return')
plt.plot(return_521_551 / return_521_551[0], label = 'S&P 500', color = 'red')
plt.plot(x_521_551 / x_521_551[0], label = 'portfolio', color = 'blue')
plt.legend()
plt.show()
plt.savefig('521-551.png')
```
